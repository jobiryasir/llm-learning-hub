# 01 - Prompt Patterns

Zero-shot, few-shot, system prompts, role-based prompts, structured output, and prompt templates.


```python
import numpy as np
import re
import matplotlib.pyplot as plt
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

print('Libraries loaded!')
```

## 1. Zero-Shot Prompting


```python
# Zero-shot: No examples, just instruction
def zero_shot_prompt(task, input_text):
    return f'Task: {task}\nInput: {input_text}\nOutput:'

examples = [
    ('Classify sentiment', 'I love this product!'),
    ('Extract keywords', 'Machine learning is transforming industries worldwide.')
]
for task, inp in examples:
    prompt = zero_shot_prompt(task, inp)
    print(f'Prompt:\n{prompt}\n')
```

## 2. Few-Shot Prompting


```python
# Few-shot: Provide examples before the actual task
def few_shot_prompt(task, examples, input_text):
    prompt = f'Task: {task}\n\n'
    for ex_input, ex_output in examples:
        prompt += f'Input: {ex_input}\nOutput: {ex_output}\n\n'
    prompt += f'Input: {input_text}\nOutput:'
    return prompt

sentiment_examples = [
    ('This movie was terrible', 'Negative'),
    ('Best film I have ever seen', 'Positive'),
    ('It was okay, nothing special', 'Neutral')
]
prompt = few_shot_prompt('Classify sentiment', sentiment_examples, 'I really enjoyed the concert!')
print(prompt)
```

## 3. System Prompts


```python
# System prompt sets the behavior/context
SYSTEM_PROMPTS = {
    'helpful_assistant': 'You are a helpful AI assistant. Answer concisely and accurately.',
    'expert_programmer': 'You are an expert Python programmer. Provide optimized, well-documented code.',
    'creative_writer': 'You are a creative writer. Use vivid imagery and engaging narratives.',
    'skeptical_analyst': 'You are a skeptical analyst. Question assumptions and verify claims.'
}

def format_with_system(system_prompt, user_message):
    return f'[SYSTEM] {system_prompt}\n\n[USER] {user_message}\n\n[ASSISTANT]'

for name, prompt in SYSTEM_PROMPTS.items():
    formatted = format_with_system(prompt, 'Explain recursion')
    print(f'\n=== {name} ===')
    print(formatted[:150] + '...')
```

## 4. Role-Based Prompts


```python
# Different roles produce different outputs
ROLES = {
    'teacher': 'Explain as if teaching a beginner student. Use simple analogies.',
    'researcher': 'Provide a detailed, academic explanation with citations.',
    'developer': 'Give a practical, code-focused explanation with examples.',
    'executive': 'Provide a high-level summary focused on business impact.'
}

def role_prompt(role, topic):
    return f'You are a {role}. {ROLES[role]}\n\nExplain: {topic}'

topic = 'Neural Networks'
for role in ROLES:
    print(f'\n=== {role.upper()} ===')
    print(role_prompt(role, topic)[:200] + '...')
```

## 5. Structured Output


```python
# Prompt for structured JSON output
def structured_prompt(entity):
    lines = [
        f'Extract information about {entity} and return JSON:',
        '{',
        '  "name": "",',
        '  "type": "",',
        '  "attributes": [],',
        '  "relationships": []',
        '}'
    ]
    return '\n'.join(lines)

print(structured_prompt('Machine Learning'))

mock_output = {
    'name': 'Machine Learning',
    'type': 'Field of AI',
    'attributes': ['Data-driven', 'Pattern recognition', 'Predictive modeling'],
    'relationships': ['Artificial Intelligence', 'Statistics', 'Computer Science']
}
print(f'\nParsed output: {mock_output}')
```

## 6. Prompt Templates


```python
class PromptTemplate:
    def __init__(self, template):
        self.template = template
        self.variables = re.findall(r'\{(\w+)\}', template)
    def format(self, **kwargs):
        return self.template.format(**kwargs)

templates = {
    'summarize': 'Summarize the following text in {max_words} words:\n\n{text}',
    'translate': 'Translate the following from {source_lang} to {target_lang}:\n\n{text}',
    'qa': 'Context: {context}\n\nQuestion: {question}\n\nAnswer:',
    'code': 'Write a {language} function that {description}. Include docstrings and error handling.'
}

for name, tmpl in templates.items():
    pt = PromptTemplate(tmpl)
    print(f'\n=== {name} ===')
    print(f'Variables: {pt.variables}')

qa_template = PromptTemplate(templates['qa'])
prompt = qa_template.format(context='The sky is blue.', question='What color is the sky?')
print(f'\nExample QA prompt:\n{prompt}')
```

## 7. Prompt Effectiveness Comparison


```python
prompt_types = ['Zero-Shot', 'Few-Shot (1)', 'Few-Shot (3)', 'Few-Shot (5)', 'System + Few-Shot', 'Chain-of-Thought']
accuracy = [45, 62, 78, 82, 85, 91]
consistency = [30, 55, 70, 75, 80, 88]

x = np.arange(len(prompt_types))
width = 0.35

plt.figure(figsize=(12, 5))
plt.bar(x - width/2, accuracy, width, label='Accuracy', color='steelblue')
plt.bar(x + width/2, consistency, width, label='Consistency', color='coral')
plt.xlabel('Prompt Type')
plt.ylabel('Score (%)')
plt.title('Prompt Pattern Effectiveness Comparison')
plt.xticks(x, prompt_types, rotation=15, ha='right')
plt.legend()
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
